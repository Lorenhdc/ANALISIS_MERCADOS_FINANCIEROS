```python
import pandas as pd
import numpy as np
```

```python
def crear_columnas(df,nombre,sector,tipo,area):
    # Crear columnas
    df.insert(1, "NOMBRE", nombre, allow_duplicates=False)
    df.rename(columns={'Price':'PRECIO'},inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.iloc[:,0:3]
    df['PRECIO NORMALIADO MAX'] = df['PRECIO']*1000/df['PRECIO'].max()
    df['PRECIO NORMALIADO MIN'] = df['PRECIO']*1/df['PRECIO'].min()
    df.insert(2, "SECTOR", sector, allow_duplicates=False)
    df.insert(2, "TIPO", tipo, allow_duplicates=False)
    df.insert(2, "AREA", area, allow_duplicates=False)

    #Crear rentabilidades

    lista_precios_semanal = []
    lista_precios_mensual = []
    lista_precios_anual = []
    lista_precios_bianual = []
    lista_precios_lustro = []
    
    for x,i in enumerate (df['PRECIO']):
        if x + 1 >= len(df['PRECIO']):
            lista_precios_semanal.append(np.nan)
        else: 
            lista_precios_semanal.append(df['PRECIO'][x+1])
    df['PRECIO SEMANAL']=lista_precios_semanal
    df['RENTABILIDAD SEMANAL']=(df['PRECIO']-df['PRECIO SEMANAL'])/df['PRECIO SEMANAL']

    for x,i in enumerate (df['PRECIO']):
        if x + 4 >= len(df['PRECIO']):
            lista_precios_mensual.append(np.nan)
        else: 
            lista_precios_mensual.append(df['PRECIO'][x+4])
    df['PRECIO MENSUAL']=lista_precios_mensual
    df['RENTABILIDAD MENSUAL']=(df['PRECIO']-df['PRECIO MENSUAL'])/df['PRECIO MENSUAL']

    for x,i in enumerate (df['PRECIO']):
        if x + 52 >= len(df['PRECIO']):
            lista_precios_anual.append(np.nan)
        else: 
            lista_precios_anual.append(df['PRECIO'][x+52])
    df['PRECIO ANUAL']=lista_precios_anual
    df['RENTABILIDAD ANUAL']=(df['PRECIO']-df['PRECIO ANUAL'])/df['PRECIO ANUAL']

    for x,i in enumerate (df['PRECIO']):
        if x + 104 >= len(df['PRECIO']):
            lista_precios_bianual.append(np.nan)
        else: 
            lista_precios_bianual.append(df['PRECIO'][x+104])
    df['PRECIO BIANUAL']=lista_precios_bianual
    df['RENTABILIDAD BIANUAL']=(df['PRECIO']-df['PRECIO BIANUAL'])/df['PRECIO BIANUAL']

    for x,i in enumerate (df['PRECIO']):
        if x + 261 >= len(df['PRECIO']):
            lista_precios_lustro.append(np.nan)
        else: 
            lista_precios_lustro.append(df['PRECIO'][x+261])
    df['PRECIO LUSTRO']=lista_precios_lustro
    df['RENTABILIDAD LUSTRO']=(df['PRECIO']-df['PRECIO LUSTRO'])/df['PRECIO LUSTRO']

    # Rentabilidades a posteriori

    lista_precios_asemestral = []
    lista_precios_aanual = []
    lista_precios_abianual = []
    lista_precios_alustro = []

    
    for x,i in enumerate (df['PRECIO']):
        if x - 24 < 0:
            lista_precios_asemestral.append(np.nan)
        else: 
            lista_precios_asemestral.append(df['PRECIO'][x-1])
    df['PRECIO SEMESTRAL A']=lista_precios_asemestral
    df['A 6 MESES']=(df['PRECIO SEMESTRAL A']-df['PRECIO'])/df['PRECIO']

    for x,i in enumerate (df['PRECIO']):
        if x - 52 < 0:
            lista_precios_aanual.append(np.nan)
        else: 
            lista_precios_aanual.append(df['PRECIO'][x-52])
    df['PRECIO ANIO A']=lista_precios_aanual
    df['A 1 ANIO']=(df['PRECIO ANIO A']-df['PRECIO'])/df['PRECIO']
    
    for x,i in enumerate (df['PRECIO']):
        if x - 104 < 0:
            lista_precios_abianual.append(np.nan)
        else: 
            lista_precios_abianual.append(df['PRECIO'][x-104])
    df['PRECIO 2 ANIO A']=lista_precios_abianual
    df['A 2 ANIOS']=(df['PRECIO 2 ANIO A']-df['PRECIO'])/df['PRECIO']

    for x,i in enumerate (df['PRECIO']):
        if x - 261 < 0:
            lista_precios_alustro.append(np.nan)
        else: 
            lista_precios_alustro.append(df['PRECIO'][x-261])
    df['PRECIO LUSTRO A']=lista_precios_alustro
    df['A 5 ANIOS']=(df['PRECIO LUSTRO A']-df['PRECIO'])/df['PRECIO']

    df.to_csv(F"COMPLETADOS/{nombre}.csv", index = False)
    

    return df
```
